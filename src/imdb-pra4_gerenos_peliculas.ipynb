Extracción de dos ficheros csv, agrupadas por genero de peliculas y el año de estreno como sus totales del genero por año

```python
titles  =  pd.read_table("title.basics.tsv", low_memory=False)
ratings  =  pd.read_table("title.ratings.tsv", low_memory=False)
```

```python
vals = titles[(titles['titleType']=='movie') & (titles['genres']!='\\N')]
# vals.head()
```

```python
all_rec_df = [pd.Series(row['startYear'],row['genres'].split(","))              
                    for _, row in vals.iterrows()]
all_rec_df = pd.concat(all_rec_df).reset_index()
```

```python
all_rec_df.columns = ['genre','year']
all_rec_df['year'] = all_rec_df['year'].astype('str')
all_genres = all_rec_df
```

```python
year_genre = pd.pivot_table(all_genres,  index=['year'], columns=['genre'], aggfunc= len,fill_value=0)
```

```python
year_genre['Total'] = year_genre.sum(axis = 1)
year_genre = year_genre.drop('\\N')
# year_genre
```

```python
year_genre.to_csv("year_genre.csv")
```
